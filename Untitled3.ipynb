```python
# Step-by-step code for Grandma's Voice System
# 1. Record a voice sample (minimum 30 seconds in WAV format)
# 2. Upload a PDF containing a story
# 3. Clone the voice and synthesize the story

# Requirements:
# pip install TTS PyPDF2 soundfile

import os
from datetime import datetime
import PyPDF2
from TTS.api import TTS

# ------------------------------------------------------
# 1. Extract text from a PDF story file
# ------------------------------------------------------
def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text
    return text

# ------------------------------------------------------
# 2. Load a multi-speaker, multilingual TTS model (YourTTS)
# ------------------------------------------------------
tts = TTS(model_name="tts_models/multilingual/multi-dataset/your_tts", progress_bar=True)

# ------------------------------------------------------
# 3. Synthesize audio using cloned voice
# ------------------------------------------------------
def synthesize_with_cloned_voice(text, reference_audio_path, output_dir="output_audio"):
    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    # Create a unique output file name using timestamp
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    output_path = os.path.join(output_dir, f"grandma_story_{timestamp}.wav")

    # Generate the audio
    tts.tts_to_file(
        text=text,
        speaker_wav=reference_audio_path,
        language="en",  # Required for multilingual models
        file_path=output_path
    )

    print(f"✅ Audio saved to: {output_path}")
    return output_path

# ------------------------------------------------------
# 4. Usage: Provide paths and run the process
# ------------------------------------------------------
pdf_path = "/content/story.pdf"  # Upload your story as PDF
reference_audio_path = "/content/trump-wef-2018_Y6q962Kr.wav"  # Upload your 30-sec WAV voice sample

# Extract story text and synthesize audio
story_text = extract_text_from_pdf(pdf_path)
output_audio_path = synthesize_with_cloned_voice(story_text, reference_audio_path)
```
